````python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
sns.set_palette(['#00A99D', '#F5CA0C', '#B6129f', '#76620C', '#095C57'])
from scipy import stats
````

````python
twitter_mu = 208.0
twitter_sigma = 60.0
twitter_followers = np.random.normal(twitter_mu, twitter_sigma, 10000)


facebook_mu = 190.0
facebook_sigma = 36.0
facebook_friends = np.random.normal(facebook_mu, facebook_sigma, 10000)
````

````python
print(twitter_followers[:12])
print('\n')
print(facebook_friends[:12])
````

### Normal Distribution 

````python
sns.distplot(twitter_followers, label = 'Twitter Followers')
sns.distplot(facebook_friends, label = 'Facebook Friends')
c= plt.legend()
````

````python
c = sns.boxplot([twitter_followers, facebook_friends], order= ['Twitter Followers', 'Facebook Friends'])
````

````python
z_score_twitter_followers = stats.mstats.zscore(twitter_followers)
z_score_facebook_friends = stats.mstats.zscore(facebook_friends)
````

````python
print('z_score_facebook_friends : \n', z_score_facebook_friends[:10])
print('\nz_score_twitter_followers : \n',z_score_twitter_followers[:10])
````

````python
# Plot of the normalized distriubutions
sns.distplot(z_score_twitter_followers, label = 'Twitter followers')
sns.distplot(z_score_facebook_friends, label = 'Facebook followers')
c = plt.legend(title = 'Normalized values (z - Scores)')
````

````python
c = sns.boxplot([z_score_twitter_followers, z_score_facebook_friends], order = ['Twitter Followers', 'Facebook Followers'])
````

### Maping to other distributions

#### What Is a Z-Score?
``` A Z-score is a numerical measurement used in statistics of a value's relationship to the mean (average) of a group of values, measured in terms of standard deviations from the mean. If a Z-score is 0, it indicates that the data point's score is identical to the mean score. A Z-score of 1.0 would indicate a value that is one standard deviation from the mean. Z-scores may be positive or negative, with a positive value indicating the score is above the mean and a negative score indicating it is below the mean. For more details please visit ```https://www.investopedia.com/terms/z/zscore.asp



````python
me_twitter_followers = 148.0
me_twitter_z_score = (me_twitter_followers - twitter_mu)/twitter_sigma
print(me_twitter_z_score)
````

````python
me_facebook_friends = me_twitter_z_score*facebook_sigma + facebook_mu
print(me_facebook_friends)
````
