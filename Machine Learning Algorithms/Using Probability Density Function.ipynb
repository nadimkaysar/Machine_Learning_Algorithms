````python
from __future__ import division
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats
````

````python
facebook_mu = 190.0
facebook_sigma = 36.0
facebook_friends = np.random.normal(facebook_mu, facebook_sigma, 10000)
````

### What is Probability Density Function (PDF)
``` In probability theory, a probability density function, or density of a continuous random variable, is a function whose value at any given sample in the sample space can be interpreted as providing a relative likelihood that the value of the random variable would equal that sample.```

````python
x = 250.0
sns.distplot(facebook_friends, label = 'Facebook Friends')
plt.text(x+5, 0.0003, '$x$='+str(x))

x_plot = np.linspace(min(facebook_friends), x, 1000)
y_plot = stats.norm.pdf(x_plot, facebook_mu, facebook_sigma)
plt.fill_between(x_plot, y_plot)
c = plt.legend()
````

````python
z_score = (x - facebook_mu)/facebook_sigma
print(z_score)
````

````python
probability = stats.norm.cdf(z_score)
print(probability)
````

````python
#  The probability  of having more than 250 people
1-probability
````

````python
# Let's assume we have a 20% o chance of having a certain number of Facebook friends or more. What is the minimum
# number of Facebook friends we have in this case
````

````python
probability = 1 - .20
z = stats.norm.ppf(probability)
print(z)
````

````python
x = z * facebook_sigma + facebook_mu
print(x)
````

````python
sns.distplot(facebook_friends, label = 'Facebook Friends', kde = False, fit = stats.norm, color = 'w')
plt.text(x+5, 0.0003, '$x$='+str(int(x)))
x_plot = np.linspace(x, max(facebook_friends), 1000)
y_plot = stats.norm.pdf(x_plot, facebook_mu, facebook_sigma)
plt.fill_between(x_plot, y_plot)
c = plt.legend()
````

````python
x1 = 120.0
x2 = 170.0

sns.distplot(facebook_friends, label = 'Facebook Friends', kde = False,
             fit = stats.norm, color = 'b')
plt.text(x1+5, 0.0003, '$x_1$='+str(int(x1)))
plt.text(x2+5, 0.0003, '$x_2$='+str(int(x2)))
x_plot = np.linspace(x1, x2, 1000)
y_plot = stats.norm.pdf(x_plot, facebook_mu,facebook_sigma)
plt.fill_between(x_plot, y_plot)
c = plt.legend()
````

````python
z1= (x1 - facebook_mu)/facebook_sigma
z2 = (x2 - facebook_mu)/facebook_sigma
print(z1)
print('\n', z2)
````

````python
probability1 = stats.norm.cdf(z1)
probability2 = stats.norm.cdf(z2)
print(probability1)
print('\n', probability2)
````

````python
final_probability = probability2 - probability1
print(final_probability)
````
