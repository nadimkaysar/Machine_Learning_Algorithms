```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = np.linspace(0, 10, 50)
print(x[-20:])
```

```python
print(plt.style.available)
```

```python
plt.title('Default Matplotlib')
plt.xlabel('X - axis')
plt.ylabel('Y - axis')
plt.plot(x, x**2)
plt.grid(True)
plt.show()
```

```python
plt.style.use('bmh')
plt.title('Baysian Method for hakers')
plt.xlabel('X - axis')
plt.ylabel('Y -axis')
plt.plot(x, x**3)
plt.grid(True)
plt.show()
```

```python
plt.style.use('ggplot')
plt.title('GGplot')
plt.xlabel('X - axis')
plt.ylabel('Y - axis')
plt.plot(x, x**3)
plt.show()
```

```python
plt.style.use('grayscale')
plt.title('Gray scale')
plt.xlabel('X - axis')
plt.ylabel('Y - axis')
plt.plot(x, x**3)
plt.grid(True)
plt.show()
```

```python
plt.style.use('dark_background')
plt.title('Dark back ground')
plt.xlabel('X axis')
plt.ylabel('Y - axis')
plt.plot(x, x**3)
plt.grid(True)
plt.show()
```
