```python
buket_size = 1000
buker_calculator = lambda x: int(x) / buket_size*buket_size
```

```python
data['bucket'] = data.steps.apply(buker_calculator)
buket_min = data.bucket.min()
buket_max = data.bucket.max()
bins = int((buket_max - buket_min)/buket_size)

data.bucket.hist(color = 'b', alpha = 0.7, figsize=(8, 10), bins = bins)
print('Mode : ', data.groupby('bucket').steps.count().idxmax(), \
     'with buket size ', buket_size)
```
